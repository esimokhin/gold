## **Описание проекта**

### **Постановка задачи**

В проекте описан прототип модели машинного обучения для компании «Цифра».<br />
Данная модель предсказывает коэффициент восстановления золота из золотосодержащей руды на основании данных с параметрами добычи и очистки.<br />
Представленная модель предназначена для оптимизации производства и позволяет избежать запуска предприятия с убыточными характеристиками.<br />



### **Технологический процесс**

После первичной обработки добытой руды получается дроблёная смесь. Её отправляют на флотацию (обогащение) и двухэтапную очистку.<br />
1. Флотация
Во флотационную установку подаётся смесь золотосодержащей руды. После обогащения получается черновой концентрат и «отвальные хвосты», то есть остатки продукта с низкой концентрацией ценных металлов.
На стабильность этого процесса влияет непостоянное и неоптимальное физико-химическое состояние флотационной пульпы (смеси твёрдых частиц и жидкости).
2. Очистка 
Черновой концентрат проходит две очистки. На выходе получается финальный концентрат и новые отвальные хвосты.

### **Описание данных**

#### Технологический процесс
- Rougher feed — исходное сырье
- Rougher additions (или reagent additions) — флотационные реагенты: Xanthate, Sulphate, Depressant
    - Xanthate — ксантогенат (промотер, или активатор флотации);
    - Sulphate — сульфат (на данном производстве сульфид натрия);
    - Depressant — депрессант (силикат натрия).
- Rougher process (англ. «грубый процесс») — флотация
- Rougher tails — отвальные хвосты
- Float banks — флотационная установка
- Cleaner process — очистка
- Rougher Au — черновой концентрат золота
- Final Au — финальный концентрат золота

#### Параметры этапов
- air amount — объём воздуха
- fluid levels — уровень жидкости
- feed size — размер гранул сырья
- feed rate — скорость подачи

### **Заполнение пропущенных данных**

В связи с тем, что процесс обогащения золота непрерывен, записи упорядочены по времени и значения параметров меняются постепенно, целесообразно заполнять пропуски параметров скользящим средним предыдущих по времени их значений. 
Оставшиеся после этого пропуски заполняем медианными значениями параметров

```python
def rolling_mean_meadian(dff, window = 10, min_periods = 1):
    print ('Пропущенных данных до обработки:', dff.isna().sum().sum())
    for col in dff.columns[1:]:
        dff['rolling'] = dff[col].rolling(window, min_periods =  min_periods).mean()
        dff[col] = dff[col].fillna(dff['rolling'])
        dff[col] = dff[col].fillna(dff[col].median())
    print ('Пропущенных данных после обработки:', dff.isna().sum().sum())    
    return dff
```

### **Сравнение распределения размеров гранул сырья на обучающей и тестовой выборках** <br />

Если распределения сильно отличаются друг от друга, оценка модели будет неправильной

```python
def calc_p_value(s_train, s_test):
    print('\033[1;3m Гипотезы:\033[0m')
    print('Нулевая гипотеза: "Распределения размеров гранул сырья на обучающей и тестовой выборках одинаковые"')
    print('Альтернативная гипотеза: "Распределения размеров гранул сырья на обучающей и тестовой выборках разные"')
    print('\n\033[1;3m Обучающая выборка \033[0m')
    train_mean = s_train.describe()['mean']
    print('Среднее значение выборки: {:.2f}'.format(train_mean))
    train_std = s_train.describe()['std']
    print('Стандартное отклонение: {:.2f}'.format(train_std))
    print('\nОписание данных:')
    display(s_train.describe())
    print('\033[1;3m Тестовая выборка \033[0m')
    test_mean = s_test.describe()['mean']
    print('Среднее значение выборки: {:.2f}'.format(test_mean))
    test_std = s_test.describe()['std']
    print('Стандартное отклонение:{:.2f}'.format(test_std))
    print('\nОписание данных:')
    display(s_test.describe())
    
    #xmin = min(s_train.quantile(q=0.001),s_test.quantile(q=0.001))
    xmax = max(s_train.quantile(q=0.995),s_test.quantile(q=0.995))
    xmin = min(s_train.describe()['min'], s_test.describe()['min'])
    #xmax = max(s_train.describe()['max'], s_test.describe()['max'])
    accuracy = 0.1
    xpoints = int((xmax-xmin)/accuracy + 1)

    # Графики:
    
    #Данные для оси Х
    grafic_x = [ round(x, 1) for x in np.linspace(xmin, xmax, xpoints)]

    fig, (ax_1, ax_2, ax_3) = plt.subplots(nrows=3, ncols=1, figsize=(12,8), sharex='all',sharey='all')
    plt.xlim(xmin, xmax)
    ax_1.set(title = 'Обучающая выборка')
    ax_2.set(title = 'Тестовая выборка')
    ax_3.set(title = 'Сравнение')

    #Гистограммы
    s_train.plot(kind = 'hist',ax=ax_1, bins=50, density=True, alpha=0.6, color='b', grid=True)
    s_test.plot(kind = 'hist',ax=ax_2, bins=50, density=True, alpha=0.6, color='g', grid=True)
    s_test.plot(kind = 'hist',ax=ax_3, bins=50, density=True, alpha=0.6, color='w', grid=True)

    #Графики
    train_p = norm.pdf(grafic_x, train_mean, train_std)
    ax_1.plot(grafic_x, train_p, linewidth=2.0, color='r')
    ax_3.plot(grafic_x, train_p, linewidth=2.0, color='r')

    test_p = norm.pdf(grafic_x, test_mean, test_std)
    ax_2.plot(grafic_x, test_p, linewidth=2.0, color='b')
    ax_3.plot(grafic_x, test_p, linewidth=2.0, color='b')
    plt.show()
    
    #Дисперсии:
    disp_train = train_std**2
    disp_test = test_std**2
    
    # T-тест Стьюдента
    if abs((disp_train-disp_test)/disp_train) > 0.05:
        eq_var = False
    else:
        eq_var = True

    result = st.ttest_ind(features_train['rougher.input.feed_size'], features_test['rougher.input.feed_size'], equal_var=eq_var)
    print('p-value = {:.4f}'.format(result.pvalue))
    if result.pvalue < ALPHA:
        print('Нулевая гипотеза отвергнута, принимается альтернативная гипотеза')
    else:    
        print('Нулевая гипотеза принимается')
```

```python
calc_p_value(features_train['rougher.input.feed_size'],features_test['rougher.input.feed_size'])
```

#### Вспомогательные функции

```python
def corr_heatmap(df, title='Корелляция данных',width = 12, height = 5):    #Построение тепловой карты корелляции
    plt.figure(figsize=(width, height))
    # define the mask to set the values in the upper triangle to True
    mask = np.triu(np.ones_like(df.corr(), dtype=np.bool))
    heatmap = sns.heatmap(df.corr(), mask=mask, vmin=-1, vmax=1, annot=True, cmap='BrBG')
    heatmap.set_title(title, fontdict={'fontsize':12}, pad=16);
    plt.show()
```

```python
def corr_killer(df,trh):  #Вычисление и удаление столбцов с высокой корелляцией
    max=1
    while max > trh:
        data_corr=df.corr()
        for col in df.columns:
            data_corr[col][col] = 0
        data_corr = data_corr.abs()
        sort_corr = data_corr.max().sort_values(ascending=False).reset_index()
        max = sort_corr[0][0]
        if max > trh:
            dead_col = sort_corr['index'][0]
            df=df.drop(dead_col, axis=1)
            
    return df.columns
```

```python
stage_cols = {} #Столбцы по стадиям
good_stage_cols = {}
good_columns = [] #Итоговый список столбцов
for i in range(len(stages)):
    stage = stages[i]
    print('Стадия \033[1;3m' +  stage + '\033[0m')
    stage_cols[stage] = []
    good_stage_cols[stage] = []
    for col in features_test.columns:
        if stage in col:
            stage_cols[stage].append(col)
    good_stage_cols[stage] = list(corr_killer(features_train[stage_cols[stage]],CORR_TRH))
    good_columns += good_stage_cols[stage]
    if len(good_stage_cols[stage]) < len(stage_cols[stage]):
        corr_heatmap(features_train[stage_cols[stage]],'Кореляция до оптимизации')
        corr_heatmap(features_train[good_stage_cols[stage]],'Кореляция после оптимизации')
    else:
        corr_heatmap(features_train[stage_cols[stage]],'Корелляция')
        print('Оптимизация не требуется')
    print('-------------------------------------------------------------------------------------------------------------')
print('\033[1;3m Все стадии после оптимизации  \033[0m')
super_good = list(corr_killer(features_train[good_columns],CORR_TRH))
if len(super_good) < len(good_columns):
    corr_heatmap(features_train[good_columns],'Кореляция до оптимизации', width = 16, height = 16)
    corr_heatmap(features_train[super_good],'Кореляция после оптимизации', width = 16, height = 16)
else:
    corr_heatmap(features_train[good_columns],'Корелляция', width = 16, height = 16)
    print('Оптимизация не требуется')
```

#### Функция для вычисления sMAPE

```python
def smape(target, predict):
    return -1*(1/len(target) * np.sum(2 * np.abs(predict-target) / (np.abs(target) + np.abs(predict))*100))
smape_scorer = make_scorer(smape, greater_is_better=False)
```

#### Функция дополнения sMAPE до 100%<br />
Используется в качестве метрики в алгоритмах, ищущих результат с наибольшей метрикой

```python
def smape2(target, predict):
    return 1*(100-(1/len(target) * np.sum(2 * np.abs(predict-target) / (np.abs(target) + np.abs(predict))*100)))
smape2_scorer = make_scorer(smape2, greater_is_better=True)
```

#### Функция для вычисления итогового sMAPE

```python
def final_smape(rougher, final):
    return 0.25 * rougher + 0.75 * final
```

#### Функция отображения важности признаков в модели (для "дерева" и "леса")

```python
def feature_importances(model, columns): 
    feats = {}
    for feature, importance in zip(columns, model.feature_importances_):
        feats[feature] = importance
    importances = pd.DataFrame.from_dict(feats, orient='index').rename(columns={0: 'Важность'})
    importances = importances.sort_values(by='Важность', ascending=False)
    importances = importances.reset_index()
    importances = importances.rename(columns={'index': 'Признаки'})
    sns.set(font_scale = 3)
    sns.set(style="whitegrid", color_codes=True, font_scale = 1.2)
    fig, ax = plt.subplots()
    fig.figsize=(12, 12)
    #fig.set_size(12,6)
    sns.barplot(x=importances['Важность'], y=importances['Признаки'], data=importances, color='skyblue')
    plt.xlabel('Важность', fontsize=14, weight = 'bold')
    plt.ylabel('Признаки', fontsize=14, weight = 'bold')
    plt.title('Важность признаков в модели', fontsize=16, weight = 'bold')
    plt.show()
    #display(plt.show())
    display(importances)
```
